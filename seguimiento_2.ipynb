```python
import numpy as np 
import pandas as pd 
import scipy.stats as stats
from statsmodels.stats.proportion import proportions_ztest

df = pd.read_excel('diabetes.xlsx')

df.head()
```

**1. ¿El valor promedio de colesterol total de la muestra, correspondiente a pacientes masculinos es menor 200mg/dL? Y en las pacientes femeninas, ¿es menor 200mg/dL?**

R/ Se realiza con una prueba t para una sola muestra. El parámetro alternative hay que ponerle less. s4 (colesterol total) hay que separarlo por generos

H0 -> µ >= 200

H1 -> µ < 200


```python
hombres = df[df["SEX"] == 2]["S1"] 
mujeres = df[df["SEX"] == 1]["S1"]

#normalidad 
#H 
stat_h, p_h_norm = stats.shapiro(hombres)

#M
stat_m, p_m_norm = stats.shapiro(mujeres)

print("Hombres normalidad p =", p_h_norm)
print("Mujeres normalidad p =", p_m_norm)
```

p > 0.05 por lo tanto usamos prueba t

```python
mu = 200

t_h, p_h = stats.ttest_1samp(hombres, mu)
t_m, p_m = stats.ttest_1samp(mujeres, mu)

p_h = p_h / 2
p_m = p_m / 2

print("Hombres p =", p_h)
print("Mujeres p =", p_m)
```

**2. ¿Hay diferencias entre el nivel de colesterol ldl entre pacientes con menos
de 40 años y más de 40 años de edad?**


R/ Se debe realizar una prueba t de comparaciones multiples.

H0 -> µ1 = µ2

H1 -> µ1 ≠ µ2

```python
menor40 = df[df["AGE"] < 40]["S2"]
mayor40 = df[df["AGE"] >= 40]["S2"]
stat1, p1 = stats.shapiro(menor40)
stat2, p2 = stats.shapiro(mayor40)

print("Menor40 p =", p1)
print("Mayor40 p =", p2)
```

```python
stat_levene, p_levene = stats.levene(menor40, mayor40)

print("Levene p =", p_levene)
```
